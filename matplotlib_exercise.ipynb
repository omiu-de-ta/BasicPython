## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df=pd.read_csv('data/hist_data.csv')
plt.hist(df['aデータ'],bins=30,color='y',alpha=0.5,label='aデータ')
plt.hist(df['bデータ'],bins=30,color='c',alpha=0.5,label='bデータ')

plt.title('aデータとbデータの温度に関する度数分布表')
plt.xlabel('温度(℃)')
plt.ylabel('度数')

plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2 = pd.read_csv('data/scatter_data.csv')
plt.scatter(df2[df2['class']==0]['length (cm)'] ,df2[df2['class']==0]['width (cm)'],label='class0',color='red')
plt.scatter(df2[df2['class']==1]['length (cm)'] ,df2[df2['class']==1]['width (cm)'],label="class1",color='blue')
plt.scatter(df2[df2['class']==2]['length (cm)'] ,df2[df2['class']==2]['width (cm)'],label='class2',color='green')

plt.title('クラス０～２のlenghtとwidthに関する散布図')
plt.xlabel('length(cm)')
plt.ylabel('width (cm)')

plt.legend()
plt.grid(True)
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3=pd.read_csv('data/line_data.csv')
kensuu = df3['件数']
kikan = df3['期間']
hiyou = df3['費用']

fig,axes=plt.subplots(2,1,figsize=(7, 10))

axes[0].plot(kikan,kensuu,marker='o')
axes[0].set_title('期間ごとの件数の推移')
axes[0].set_xlabel('期間')
axes[0].set_ylabel('件数')
axes[0].set_xticklabels(kikan,rotation=30)
axes[0].grid(True)


axes[1].bar(kikan,hiyou)
axes[1].set_title('期間ごとの費用の推移')
axes[1].set_xlabel('期間')
axes[1].set_ylabel('費用(円)')
axes[1].set_xticklabels(kikan,rotation=30)
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
